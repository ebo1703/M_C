```python
#Funci칩n para encontrar que intervalos tienen el 6/4 (Garantizando que sea 4 justa)


def CuartaseIntervalos4(dataframe):
  nuevalista = dataframe['Nombre intervalo simplificado']
  # info_filas = []
  indices = []
  for indice, lista in enumerate(nuevalista):
    if ('M6' in lista or 'm6' in lista) and 'P4' in lista:
      #FILAS E INFORMACION
      fila = dataframe.iloc[indice]
      # info_filas.append(fila)
      #INDICES PARA ACCEDER POR FUERA
      indice_fila = indice
      indices.append(indice_fila)

  return indices
```

```python
#Funci칩n encontrar voces dentro del 6/4 

def voces_64(df):

    """
    Funci칩n que encuentra las voces que contienen el intervalo 6/4 en un dataframe de pandas

    Parameters
    ----------
    df : dataframe
        Dataframe de pandas con informaci칩n de los acordes de la partitura

    Returns

    n_indice4 : list
        Lista con los 칤ndices de las voces que contienen el intervalo de cuarta justa
    n_indice6 : list
        Lista con los 칤ndices de las voces que contienen el intervalo de sexta mayor o menor

        
    """
    nuevalista = df['Nombre intervalo simplificado']
    n_indice4 = []
    n_indice6 = []
    a = df['Nombre intervalo simplificado'].tolist()
    for j in a:
        
            voz = j
            indice4 = voz.index('P4') +1
            
            if 'm6' in voz:
                indice6 = voz.index('m6') + 1
            if 'M6' in voz:
                indice6 = voz.index('M6') + 1
            n_indice4.append(indice4)
            n_indice6.append(indice6)
        

    return n_indice4 , n_indice6
```

```python
#Extraer nota del bajo de los acordes 

def extraer_n4_n6(dataframe,indice4,indice6):

    """ 
    Funci칩n que extrae las notas de la cuarta y sexta de los acordes 6/4 del dataframe filtrado

    Parameters

    ----------

    dataframe : DataFrame

        Dataframe filtrado de los acordes de la partitura

    indice4 : list

        Lista con los 칤ndices de las voces que contienen el intervalo de cuarta justa

    indice6 : list

        Lista con los 칤ndices de las voces que contienen el intervalo de sexta mayor o menor

    Returns

    -------

    cuartas : list

        Lista con las notas de la cuarta de los acordes 6/4


    sextas : list

        Lista con las notas de la sexta de los acordes 6/4
         
    
    """
   
    cuartas = []
    sextas = []
    for i in range(0,len(indice4)):
        not4 = dataframe.iloc[i]['Notas'][ind4[i]][:-1]
        not6 = dataframe.iloc[i]['Notas'][ind6[i]][:-1]
        cuartas.append(not4) 
        sextas.append(not6)  
        
    return cuartas, sextas 

N4sb, N6sb = extraer_n4_n6(df.iloc[lista_64],ind4,ind6)
```

```python
def notas_bajos(df,lista_indice):
    
    """ 
    Funci칩n que extrae las notas del bajo de los acordes de una lista de 칤ndices de un dataframe
    la lista es la correspondiente a los acordes 6/4
    El dataframe es el dataframe de todos los acordes de la partirura 
    que se est치 analizando. Devuelve una lista con las notas del bajo antes y despu칠s de los acordes 6/4 

    Parameters
    ----------
    df : DataFrame
        Dataframe de los acordes de la partitura

    lista_indice : list
        Lista de 칤ndices de los acordes 6/4

    Returns
    -------
    BB_n : list
        Lista con las notas del bajo de los acordes antes de los acordes 6/4

    BA_n : list
        Lista con las notas del bajo de los acordes despu칠s de los acordes 6/4


    """
    
    
    BB_n = []
    BA_n = []

    for j in lista_indice:

        BB_notas = []
        for i in range(1,5):
            NB = df.iloc[j]['Notas'][0][0]
            BB = df.iloc[j-i]['Notas'][0][0]
            
            # BB_notas.append(NB)
            BB_notas.append(BB)
            if NB != BB:
                break
        BB_n.append(BB_notas)

        BA_notas = []

        for i in range(1,5):
            NB = df.iloc[j]['Notas'][0][0]
            BA = df.iloc[j+i]['Notas'][0][0]

            # BB_notas.append(NB)
            BA_notas.append(BA)
            if NB != BA:
                break
        BA_n.append(BA_notas)

    return BB_n, BA_n

    # df.iloc[lista_64[j]-i]['Notas'][ind4[j]][0]
        

BB_n, BA_n = notas_bajos(df,lista_64)
BB_n, BA_n
```

```python
def notas_voces(df,ind4,ind6,BB_n,BA_n,lista_64):
    
    """ 
    Funci칩n que extrae las notas de las voces (4 y 6) de los acordes antes y despu칠s de los acordes 6/4
    El dataframe es el dataframe de todos los acordes de la partirura 
    que se est치 analizando. Devuelve una lista con las notas de las voces antes y despu칠s de los
    acordes 6/4 para la cuarta y la sexta

    Parameters
    ----------
    df : DataFrame
        Dataframe de los acordes de la partitura

    ind4 : list
        Lista de 칤ndices de las voces 4

    ind6 : list
        Lista de 칤ndices de las voces 6

    BB_n : list
        Lista con las notas del bajo de los acordes antes de los acordes 6/4

    BA_n : list
        Lista con las notas del bajo de los acordes despu칠s de los acordes 6/4

    lista_64 : list
        Lista de 칤ndices de los acordes 6/4

    Returns
    -------
    voces4b : list
        Lista con las notas de la cuarta antes de los acordes 6/4

    voces4a : list
        Lista con las notas de la cuarta  despu칠s de los acordes 6/4

    voces6b : list
        Lista con las notas de la sexta  antes de los acordes 6/4

    voces6a : list
        Lista con las notas de la sexta despu칠s de los acordes 6/4


        

    """

    voces4b = []
    voces4a = []
    voces6b = []
    voces6a = []


    for j in range(len(ind4)):
        notas4b = []
        notas6b = []
        Notas4 = []
        Notas6 = []

        
        
        for i in range(1,len(BB_n[j])+1):
            N4b = df.iloc[lista_64[j]-i]['Notas'][ind4[j]][:-1]
            N6b = df.iloc[lista_64[j]-i]['Notas'][ind6[j]][:-1]
            
            notas4b.append(N4b)
            notas6b.append(N6b)

            
        voces4b.append(notas4b)
        voces6b.append(notas6b)

        notas4a = []
        notas6a = []

        for i in range(1,len(BA_n[j])+1):
            N4a = df.iloc[lista_64[j]+i]['Notas'][ind4[j]][:-1]
            N6a = df.iloc[lista_64[j]+i]['Notas'][ind6[j]][:-1]
            notas4a.append(N4a)
            notas6a.append(N6a)
    
            
        voces4a.append(notas4a)
        voces6a.append(notas6a)

    return voces4b, voces4a, voces6b, voces6a
```
